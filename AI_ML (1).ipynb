#CHART 7 -

7. Plot a kde to show the density distribution of Sizes of the property.

```python
# chart 7 - visualization code

# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

# Creating a KDE plot for the 'Size' column
sns.kdeplot(df['Size'], fill=True, color='lightgreen', label='Size KDE')

# Set plot title and labels
plt.title('Kernel Density Estimate (KDE) of Size')
plt.ylabel('Density')

# Adding a legend
plt.legend()

plt.show()
```

1. Why did you pick the specific chart ?

>I chose Kernel Density Estimation because kde provides a non-parametric way to estimate the probability density function of a random variable. This helps in understanding the likelihood of different values occurring.

2. What is/are the insight(s) found from the chart ?

>Inference -
>
>1. The kde is positively skewed.
>
>2. The location of the peak or highest density region provides an estimate of the central tendency of sizes.

>3. Wider distributions suggest higher variability, while narrower distributions indicate lower variability in sizes.

#CHART 8 -

8. Plot a graph to show the maximum rent on the basis of their furnishing status

```python
# chart 8 - visualization code

# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

# Group by 'Furnishing Status' and calculating the maximum rent in each group
max_rent = df.groupby('Furnishing Status')['Rent'].max().reset_index()

# Create a bar plot
sns.barplot(x='Furnishing Status', y='Rent', data=max_rent, palette='viridis')

# Set plot title and labels
plt.title('Maximum Rent Based on Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Maximum Rent')

plt.show()
```

1. Why did you pick the specific chart ?

>I chose bar plot using seaborn as bar plot is used for comparing categorical data and it is effectively conveying the maximum rent values associated with different furnishing statuses in a clear and accessible manner.

2. What is/are the insight(s) found from the chart ?

>Inference -
>
>1. The tallest bar on the plot corresponds to the furnishing status with the highest maximum rent, providing a clear comparison.
>
>2. For business planning , identifying which furnishing statuses yields higher maximum rents may inform investment decisions or market strategies.

3. Will the gained insights help creating a positive business impact? Are there any insights that lead to negative growth? Justify with specific reason.
